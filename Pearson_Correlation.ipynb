```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# Create the DataFrame
data = {
    'Year': list(range(2010, 2025)),
    'Reddit_Sentiment': [0.20, 0.21, 0.41, 0.39, 0.07, 0.12, 0.14, 0.28, 0.17, 0.13, 0.20, 0.14, 0.22, 0.17, 0.18],
    'PubMed_Sentiment': [0.36, 0.42, 0.40, 0.45, 0.38, 0.44, 0.45, 0.43, 0.46, 0.47, 0.49, 0.49, 0.49, 0.49, 0.49]
}
combined_sentiment = pd.DataFrame(data)

# Calculate Pearson correlation coefficient
correlation, p_value = pearsonr(combined_sentiment['Reddit_Sentiment'], combined_sentiment['PubMed_Sentiment'])

# Print the correlation and p-value
print(f'Pearson correlation coefficient: {correlation}')
print(f'P-value: {p_value}')

# Plotting the sentiment trends for visualization
plt.figure(figsize=(10, 6))
plt.plot(combined_sentiment['Year'], combined_sentiment['Reddit_Sentiment'], label='Reddit Sentiment', color='blue')
plt.plot(combined_sentiment['Year'], combined_sentiment['PubMed_Sentiment'], label='PubMed Sentiment', color='orange')
plt.xlabel('Year')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Comparison: Reddit vs PubMed (2010-2024)')
plt.legend()
plt.grid(True)
plt.show()
```
